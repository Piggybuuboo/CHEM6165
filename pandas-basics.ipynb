```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
compound_df = pd.read_csv("CHEM2025-fullexample-data.csv")
```

```python
compound_df
```

```python
compound_df["Name"]
```

```python
Number_of_Carbon_Atoms_list = []
Boiling_Point_K = []

for lines_carbon in compound_df["Number of Carbon Atoms"]:
    Number_of_Carbon_Atoms_list.append(lines_carbon)

for lines_Kelvin in compound_df["Boiling_Point_K"]:
    Boiling_Point_K.append(lines_Kelvin)

print(Number_of_Carbon_Atoms_list)
print(Boiling_Point_K)
```

```python
plt.plot(Number_of_Carbon_Atoms_list, Boiling_Point_K)
plt.show()
```
